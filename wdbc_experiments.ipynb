```python
from dataload import load_data
from clustering import run_clustering_experiments
from dimension_reduction import run_dr_experiments
```

Breast Cancer Data

```python
# Breast Cancer Data
data_name = 'WDBC'

train_x, test_x, train_y, test_y = load_data(data_name)
```

```python
kmeans, em = run_clustering_experiments(train_x, test_x, train_y, data_name, True, 
                                         kmeans_n=2, em_n=2, em_cov='full')
```

```python
#PCA (n) ICA (n) KPCA (n, kernel (poly, rbf, sigmoid, cosine)) RP (n, runs)
dr_outputs = run_dr_experiments(train_x, test_x, train_y, data_name, True,
                                pca_n=15, ica_n=10, kpca_n=15, kpca_kernel='sigmoid',
                                rp_n=30, rp_runs=10)
```

```python
dr_clusters = {}
for dr_type, x_dr in dr_outputs.items():
    kmeans, em = run_clustering_experiments(x_dr['train'], x_dr['test'], train_y, f"DR_{data_name}_{dr_type}", True, 
                                         kmeans_n=2, em_n=2, em_cov='full')
    dr_clusters[dr_type] = {
        'kmeans': kmeans,
        'em': em
    }
```

```python
run_clustering_experiments(train_x, test_x, train_y, data_name, False, em_n=2, em_cov='full',kmeans_n=2)
```

```python
run_clustering_experiments(dr_outputs['PCA']['train'], dr_outputs['PCA']['test'], train_y, data_name, False, em_n=2, em_cov='full',kmeans_n=2)
```

```python
run_clustering_experiments(dr_outputs['KPCA']['train'], dr_outputs['KPCA']['test'], train_y, data_name, False, em_n=2, em_cov='full',kmeans_n=2)
```
